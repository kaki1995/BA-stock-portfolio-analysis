```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

import warnings
warnings.filterwarnings('ignore', category=UserWarning)
warnings.filterwarnings('ignore', category=FutureWarning)
```

```python
# Importing stock daily trading of portfolio from 2022-01-01 until 2023-12-31
daily_trading = pd.read_csv("Stock_Daily_Trading.csv", sep= ";")
daily_trading
```

```python
daily_trading.info()
```

```python
# Importing financial ratios of 10 companies in the portfolio from 2022-01-01 until 2023-12-31
financial_ratios = pd.read_csv("Financial Ratios_Top 10 Profitable Companies_2y.csv", sep= ",")
financial_ratios
```

```python
financial_ratios.info()
```

```python
financial_ratios.shape
```

```python
# Rename "public_date" column from financial_ratios dataset into "date"
financial_ratios = financial_ratios.rename(columns={"public_date": "date"})
```

```python
# Sort both datasets by "date" then "company name"
daily_trading_sorted_by_stock = daily_trading.sort_values(by = ['PERMNO', 'date'])
financial_ratios_sorted_by_stock = financial_ratios.sort_values(by = ['permno', 'date'])

# Merged them by date using "union" join
merged_sorted_df = pd.merge(daily_trading_sorted_by_stock, financial_ratios_sorted_by_stock, on='date', how='outer')
merged_sorted_df["date"]
```

```python
# Check for missing values in the merged dataset
missing_values = merged_sorted_df.isnull().sum()
print(missing_values)
```

```python
# Fill missing values using backward fill
columns_to_fill = [
    'pe_exi', 'pe_inc', 'dpr', 'npm', 'gpm', 'roa', 'roe', 'debt_at',
    'dltt_be', 'intcov_ratio', 'quick_ratio', 'curr_ratio', 'inv_turn', 'at_turn'
]

for column in columns_to_fill:
    merged_sorted_df[column] = merged_sorted_df[column].fillna(method='bfill')

# Drop rows with missing values in specified columns
columns_to_drop_na = ['PERMNO', 'EXCHCD', 'TICKER_x', 'COMNAM', 'PERMCO', 'BIDLO', 'ASKHI', 'PRC', 'VOL', 'RET', 'SHROUT', 'OPENPRC', 'RETX']
merged_sorted_df.dropna(subset=columns_to_drop_na, inplace=True)

# Fill missing values in 'DIVAMT' and 'NUMTRD' columns with 0
merged_sorted_df['DIVAMT'].fillna(0, inplace=True)
merged_sorted_df['NUMTRD'].fillna(0, inplace=True)

#Drop unnessary columns: gvkey, permno, adate, qdate, TICKER_y
merged_sorted_droped_df = merged_sorted_df.drop(['gvkey', 'permno', 'adate', 'qdate', 'TICKER_y'], axis=1)

# Verify if missing values are dropped
print(merged_sorted_droped_df.isnull().sum())
```

```python
# Check data types of all columns
print(merged_sorted_droped_df.dtypes)

#Convert into correct datatypes:
merged_sorted_droped_df['PERMNO'] = merged_sorted_droped_df['PERMNO'].astype(int)
merged_sorted_droped_df['date'] = pd.to_datetime(merged_sorted_droped_df['date'])

merged_sorted_droped_df.dtypes
```
